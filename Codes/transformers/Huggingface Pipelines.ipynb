```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "I love travelling, and mountains are my favourites",
    candidate_labels=["travel", "education", "sports", "construction"]
)
```

```python
generator = pipeline("text-generation")
generator("I love to travel and")
```

```python
# Model input
import torch

generator = pipeline("text-generation", model="openai-community/gpt2", torch_dtype=torch.float16)
generator("I love to travel and")
```

```python
mask = pipeline("fill-mask")
mask("I love to <mask> and europe is my favourite!", top_k=5)
```
