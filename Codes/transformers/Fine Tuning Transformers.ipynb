# Fine tuning a Transformer model

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

model = AutoModelForSequenceClassification.from_pretrained("distilbert/distilbert-base-uncased", device_map="auto", torch_dtype="auto").to("cuda")
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilbert-base-uncased")

model, tokenizer
```

## Data preparation

```python
tokenizer.encode([ds['train'][0]['text'], ds['train'][1]['text']])
```

```python
def tokenize_dataset(example):
  return tokenizer(example["text"])

ds = ds.map(tokenize_dataset, batched=True)
ds['train'][:3]
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
data_collator
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="rotten_tomato_trained",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    push_to_hub=False,
    learning_rate=2e-5,
    num_train_epochs=2
)

training_args
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=ds['train'],
    eval_dataset=ds['test'],
    tokenizer=tokenizer
)

trainer
```

# Training

```python
trainer.train()
```

## Evaluation

```python
preds = trainer.predict(ds['test'])
preds
```

```python
preds.label_ids, ds['test']['label']
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(preds.label_ids, ds['test']['label'])
```
